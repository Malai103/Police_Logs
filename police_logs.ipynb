```python
import pandas as pd
```

```python
logs_df = pd.read_excel('traffic_stops.xlsx')
```

```python
# drop the unwanted column
logs_df.drop(['violation_raw','driver_age_raw','driver_race'],axis=1,inplace=True)
```

```python
logs_df
```

```python
logs_df.info()
```

```python
# merging the stop date and stop time
logs_df['stop_datetime'] = pd.to_datetime(logs_df['stop_date'].astype(str) + ' ' + logs_df['stop_time'])
```

```python
# convert both at str type and to datetime format
logs_df['stop_datetime'] = pd.to_datetime(
    logs_df['stop_date'].astype(str) + ' ' + logs_df['stop_time'].astype(str)
)
```

```python
logs_df
```

```python
logs_df.info()
```

```python
# drop (stop_date &  stop_time) have a new column combined as stop_datetime  
logs_df.drop(['stop_date','stop_time'],axis=1,inplace=True)
```

```python
logs_df
```

```python
# # Standardize driver_gender (m=male,=female) Removes any spaces from the string values.Converts to to uppercase.
logs_df['driver_gender'] = logs_df['driver_gender'].str.strip().str.upper()
```

```python
# now mapping (M=male,F=female) using map function
driver_gender_map = {'M': 'Male', 'F': 'Female'}
logs_df['driver_gender'] = logs_df['driver_gender'].str.strip().str.upper().map(driver_gender_map)
```

```python
# change the (stop_duration) to lower case
logs_df['stop_duration'] = logs_df['stop_duration'].str.strip().str.lower()
```

```python
logs_df.isnull().sum()
```

```python
logs_df['search_type'].describe()
```

```python
# see value count of all in particular column. By default, value_counts() ignores NaN values.
# Using dropna=False tells it:“Count NaNs too, I want to know how many are missing.”
logs_df['search_type'].value_counts(dropna=False)
```

```python
# fill the missing values with (no search) replace using (The fillna() function is used to replace missing (NaN) values in a DataFrame) 
logs_df['search_type'] = logs_df['search_type'].fillna('No Search')
```

```python
logs_df.isnull().sum()
```

```python
# finaly save the cleaned data to (.csv or excel)format
logs_df.to_excel('traffic_stops_cleaned data.xlsx', index=False)
```
